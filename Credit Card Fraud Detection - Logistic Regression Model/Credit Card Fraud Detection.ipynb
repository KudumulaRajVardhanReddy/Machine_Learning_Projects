```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'creditcard.csv')
```

```python
df.head(200)
```

```python
df.tail()
```

```python
df['Class'].value_counts() #DF is highly unbalanced, cuz legit samples are significantly higher than fraudulent ones...
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]

legit_492 = legit.sample(n=492)

corrected_df = pd.concat([legit_492, fraud], axis = 0)

X = corrected_df.drop(columns='Class', axis = 1)
Y = corrected_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
model = LogisticRegression(max_iter = 10000)
model.fit(X_train, Y_train)
```

```python
train_data_prediction = model.predict(X_train)
test_data_prediction = model.predict(X_test)
```

```python
train_data_accuracy = accuracy_score(train_data_prediction, Y_train)
test_data_accuracy = accuracy_score(test_data_prediction, Y_test)
```

```python
print(train_data_accuracy)
print(test_data_accuracy)
```
